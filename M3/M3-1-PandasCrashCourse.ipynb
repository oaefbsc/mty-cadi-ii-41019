## Series

```python
# Se definen la serie de datos e índices
mydata =[33, 19, 15, 89, 11, -5, -9]
myindex = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri','Sat','Sun']

# Crea el objeto serie s1
s1 = pd.Series(mydata, index = myindex)
```

```python
type(s1)
```

```python
s1
```

```python
# Se definen la serie de datos e índices
mydicc = {'Mon': 33, 'Tue': 19, 'Wed': 15, 'Thu': 89, 'Fri': 11, 'Sat': -5, 'Sun': 9}

# Crea el objeto serie s2
s2 = pd.Series(mydicc)
```

```python
s2
```

Las series funcionan como arreglos de numpy y tienen propiedades y métodos similares a los objetos `ndarray` y otros iterables.

```python
# operaciones vectorizadas
s3 = s2 * 5
s3
```

```python
# indexado
s3[5]
```

```python
s3['Tue']
```

```python
# slice usando posición
s3[1:3]
```

```python
# slice usando las etiquetas del índice
s3['Thu':'Wed']
```

```python
# asignación de valores
s3[1] = 199
s3['Sat'] = 35
```

```python
s3
```

```python
# método sum()
s3.sum()
```

```python
s3[1:3].sum()
```

```python
# método max()
s3.max()
```

Una serie por sí misma no es un objeto iterable, usando la función `enumerate()` se puede iterar sobre la serie.

```python
# itera sobre los valores de la serie
# usando enumerate() devuelve un objeto iterable
for i, v in enumerate(s3):
    print (i, v)
```

Para iterar por llave-valor se utiliza el método `iteritems()`

```python
# itera sobre el diccionario por medio de llave y valores
for k,v in s3.iteritems():
    print (k,v)
```

## Operaciones básicas

### Conteos

```python
mlb.Year.value_counts()
```

```python
mlb.Team.value_counts()
```

### Filtros

```python
mlb.Year == 2010
```

```python
mlb[mlb.Year == 2010]
```

```python
mlb_yr2010 = mlb[mlb.Year == 2010]
mlb_yr2010
```

```python
# crear un índice alfanumérico con base en "Name"
yr2010 = mlb_yr2010.set_index('Player')
yr2010
```

### ordenamientos

```python
# ordenar por etiquetas de renglón
yr2010.sort_index().head()
```

```python
# ordenar por valores de columna usando el método sort_values()
yr2010.Salary.sort_values(ascending=False).head(20)
```

```python
# ordenar por valores de columna usando el método sort_values()
yr2010.sort_values('Salary', ascending=False).head(20)
```

```python
# ordenar los valores de varias columnas
yr2010.sort_values(['Salary', 'Team'], ascending=[False,True]).head(10)
```

```python
titanic_grouped = titanic.groupby('Pclass')
```

```python
titanic_grouped
```

El método `groupby()` devuelve un objeto iterable el cual hay que recorrer

```python
for pcls,group in titanic_gruped:
    print(pcls)
    print(group)
```

```python
# for plcs, group in titanic_grouped
# (group.sort_values(by = 'Fare', ascending=False)
# titanic_grouped_pclass =
#[:1]
# titanic_grouped_pclass = (group.sort_values(by = 'Fare', ascending=False)[:2] for plcs, group in titanic_grouped) 
titanic_grouped_pclass = (group.sort_values(by = 'Fare', ascending=False)[:1] for plcs, group in titanic_grouped) 
```

```python
type(titanic_grouped_pclass)
```

```python
# convierte el objeto generador en un DataFrame
topfares_pcls = pd.DataFrame()
for line in titanic_grouped_pclass:
    topfares_pcls = topfares_pcls.append(line)
```

```python
topfares_pcls
```
